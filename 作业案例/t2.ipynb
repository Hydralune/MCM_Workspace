```python
import pulp

# 创建问题
model = pulp.LpProblem(name="nurse-scheduling-plan2", sense=pulp.LpMinimize)

# 决策变量
x = {i: pulp.LpVariable(f"x{i}", lowBound=0, cat=pulp.LpInteger) for i in range(1, 11)}

# 目标函数：最小化所需护士总数
model += sum(x.values())

# 约束条件
# 6:00-10:00 时间段的约束
model += x[10] + x[6] >= 18  # 周一
model += x[1] + x[5] + x[9] + x[10] >= 18  # 周二
model += x[4] + x[8] + x[5] + x[9] >= 18  # 周三
model += x[3] + x[7] + x[4] + x[8] >= 18  # 周四
model += x[2] + x[6] + x[3] + x[7] >= 18  # 周五
model += x[1] + x[2] + x[6] + x[10] >= 18  # 周六
model += x[6] + x[7] + x[8] + x[9] >= 18  # 周日

# 10:00-14:00 时间段的约束
model += x[6] + x[7] + x[8] + x[9] >= 20  # 周一
model += x[1] + x[2] + x[6] + x[10] >= 20  # 周二
model += x[1] + x[5] + x[9] + x[10] >= 20  # 周三
model += x[4] + x[8] + x[5] + x[9] >= 20  # 周四
model += x[3] + x[7] + x[4] + x[8] >= 20  # 周五
model += x[2] + x[3] + x[7] + x[8] >= 20  # 周六
model += x[7] + x[8] + x[9] + x[10] >= 20  # 周日

# 14:00-18:00 时间段的约束
model += x[7] + x[8] + x[9] + x[10] >= 19  # 周一
model += x[2] + x[6] + x[3] + x[7] >= 19  # 周二
model += x[1] + x[2] + x[6] + x[10] >= 19  # 周三
model += x[1] + x[5] + x[9] + x[10] >= 19  # 周四
model += x[4] + x[8] + x[5] + x[9] >= 19  # 周五
model += x[3] + x[4] + x[8] + x[9] >= 19  # 周六
model += x[8] + x[9] + x[10] + x[6] >= 19  # 周日

# 18:00-22:00 时间段的约束
model += x[8] + x[9] + x[10] + x[6] >= 17  # 周一
model += x[3] + x[7] + x[4] + x[8] >= 17  # 周二
model += x[2] + x[6] + x[3] + x[7] >= 17  # 周三
model += x[1] + x[2] + x[6] + x[10] >= 17  # 周四
model += x[5] + x[9] + x[1] + x[10] >= 17  # 周五
model += x[4] + x[5] + x[9] + x[10] >= 17  # 周六
model += x[9] + x[10] + x[6] + x[7] >= 17  # 周日

# 22:00-6:00 时间段的约束
model += x[9] >= 12  # 周一
model += x[5] + x[9] >= 12  # 周一
model += x[4] + x[8] >= 12  # 周二
model += x[10] + x[4] >= 12  # 周二
model += x[3] + x[7] >= 12  # 周三
model += x[5] + x[3] >= 12  # 周三
model += x[2] + x[6] >= 12  # 周四
model += x[4] + x[2] >= 12  # 周四
model += x[1] + x[10] >= 12  # 周五
model += x[3] + x[1] >= 12  # 周五
model += x[5] >= 12  # 周六
model += x[2] >= 12  # 周六
model += x[10] >= 12  # 周日
model += x[1] >= 12  # 周日

# 求解问题
model.solve()

# 输出结果
print("方案2求解状态:", pulp.LpStatus[model.status])
total_nurses = 0
for i in range(1, 11):
    print(f"x{i}: {pulp.value(x[i])}")
    total_nurses += pulp.value(x[i])
print(f"方案2需要的护士总数: {total_nurses}")
```
